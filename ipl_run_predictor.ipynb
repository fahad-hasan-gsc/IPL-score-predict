```python
import pandas as pd
import numpy as np
```

```python
csv=("https://raw.githubusercontent.com/anujvyas/IPL-First-Innings-Score-Prediction-Deployment/master/ipl.csv")
```

```python
df=pd.read_csv(csv)
df.head(10)
```

```python
df.isna().sum()
```

```python
df['bowl_team'].unique()
```

```python
df['bat_team'].unique()
```

```python
teams=[
       'Royal Challengers Bangalore', 
       'Kings XI Punjab',
       'Kolkata Knight Riders', 
       'Rajasthan Royals',
       'Mumbai Indians',
       'Chennai Super Kings', 
       'Sunrisers Hyderabad',
       'Delhi Daredevils'
]
```

```python
df['venue'].unique()
```

```python
df=df[(df['bowl_team'].isin(teams)) & (df['bat_team'].isin(teams))]
df=df.drop(columns=['batsman','bowler','bowler','striker','non-striker','venue'])
```

```python
df=df[df['overs']>5.0]
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
df['date']=df['date'].dt.year
```

```python
df
```

```python
dum_data=pd.get_dummies(df,columns=['bat_team','bowl_team'])
```

```python
dum_data
```

```python
x_train=dum_data.drop(columns='total',axis=1)[dum_data['date']<=2016]
x_test=dum_data.drop(columns='total',axis=1)[dum_data['date']>=2017]
x_train=x_train.drop(columns=['date'],axis=1)
x_test=x_test.drop(columns=['date'],axis=1)
```

```python
y_train=dum_data[dum_data['date']<=2016]['total'].values
y_test=dum_data[dum_data['date']>=2017]['total'].values
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_train.info()
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```
